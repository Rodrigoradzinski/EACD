Segundo o [PNUD do Brasil](https://www.undp.org/pt/brazil/idh) (Programa das Nações Unidas para o Desenvolvimento)

> "O Índice de Desenvolvimento Humano (IDH) é uma medida resumida do progresso a longo prazo em três dimensões básicas do desenvolvimento humano: renda, educação e saúde. O objetivo da criação do IDH foi o de oferecer um contraponto a outro indicador muito utilizado, o Produto Interno Bruto (PIB) per capita, que considera apenas a dimensão econômica do desenvolvimento. Criado por Mahbub ul Haq com a colaboração do economista indiano Amartya Sen, ganhador do Prêmio Nobel de Economia de 1998, o IDH pretende ser uma medida geral e sintética que, apesar de ampliar a perspectiva sobre o desenvolvimento humano, não abrange nem esgota todos os aspectos de desenvolvimento."

Conforme apresentado no início deste notebook, um dos pilares do IDH é a escolaridade. Com base nisso, compare graficamente o tempo de escolaridade esperado de cada país da tabela *lac* em 1990.

```python
import matplotlib.pyplot as plt


lac_1990 = lac[['Country', 'Expected Years of Schooling (1990)']]


lac_1990 = lac_1990.sort_values(by='Expected Years of Schooling (1990)', ascending=False)


plt.figure(figsize=(10, 8))
plt.barh(lac_1990['Country'], lac_1990['Expected Years of Schooling (1990)'], color='skyblue')
plt.xlabel('Expectativa em anos')
plt.ylabel('País')
plt.title('Expectativa de Anos de EstudoF ormal em 1990 (Países da América Latina e Caribe)')
plt.gca().invert_yaxis()
plt.show()
```

Compare a evolução da expectativa de anos de estudos dos países da tabela *lac* do ano 1990 e do ano 1995. Utilize o gráfico de barras vertical para isso.

```python
lac_1990_1995 = lac[['Country', 'Expected Years of Schooling (1990)', 'Expected Years of Schooling (1995)']]
lac_1990_1995 = lac_1990_1995.sort_values(by='Expected Years of Schooling (1995)', ascending=False)
lac_1990_1995.set_index('Country', inplace=True)

ax = lac_1990_1995.plot(kind='bar', figsize=(12, 8))
plt.xlabel('País')
plt.ylabel('Expectativa em anos')
plt.title('Expectativa de Anos de Estudo Formal em 1990 e 1995 (Países da América Latina e Caribe)')
plt.xticks(rotation=90)
plt.legend(['1990', '1995'])
plt.show()
```

Analisando o resultado você percebeu que os rótulos das colunas  contendo as expectativas de anos de estudo estavam muito longas. Para facilitar sua vida, deixei preparado as declarações para renomear as colunas. Contudo, acabei esquecendo de atualizar o nome *lac* com essas atualizações. Complete a declaração abaixo para que as alterações sejam salvas na tabela *lac*.

```python
# Não altere as declarações a seguir! ##########################################
oldCols = lac.columns.to_list()                                                #
removes = ['ISO3', 'Country', 'Human Development Groups',                      #
          'UNDP Developing Regions', 'HDI Rank (2021)']                        #
                                                                               #
for remove in removes:                                                         #
    oldCols.remove(remove)                                                     #
                                                                               #
newCols = [str(year) for year in range(1990, 2022)]                            #
################################################################################
# Faça as alterações necessárias na declaração abaixo para que o nome lac seja #
# atualizado com as colunas renomeadas:                                        #
################################################################################
lac = lac.rename(columns=dict(zip(oldCols, newCols)))

# A declaração abaixo irá mostrar a tabela atualizada. #########################
lac
```

Plote o gráfico de barras vertical e verifique a evolução da expectativas de anos de estudos dos anos 1990, 2000, 2010, 2020.

```python
import matplotlib.pyplot as plt


years_of_interest = ['1990', '2000', '2010', '2020']
lac_years = lac[['Country'] + years_of_interest]


lac_years.set_index('Country', inplace=True)


lac_years = lac_years.T


ax = lac_years.plot(kind='bar', figsize=(12, 8))
plt.xlabel('Ano')
plt.ylabel('Expectativa em anos')
plt.title('Evolução da Expectativa de Anos de Estudo Formal (Países da América Latina e Caribe)')
plt.legend(title='País', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=0)
plt.show()
```

Analisando o resultado, você achou melhor o gráfico debarras horizontal. Além disso, acrescentou grades e inseriu:

* título: Expectativa de Anos de Estudo Formal
* nome ao eixo horizontal: Expectativa em anos
* nome do eixo vertical: País

e alterou as dimensões da figura para que ficasse com 13 polegas de largura e 8 de altura.

```python
import matplotlib.pyplot as plt


years_of_interest = ['1990', '2000', '2010', '2020']
lac_years = lac[['Country'] + years_of_interest]


lac_years.set_index('Country', inplace=True)


lac_years = lac_years.T


ax = lac_years.plot(kind='barh', figsize=(13, 8), grid=True)
plt.xlabel('Expectativa em anos')
plt.ylabel('País')
plt.title('Expectativa de Anos de Estudo Formal')
plt.legend(title='Ano', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.gca().invert_yaxis()
plt.show()
```

Analisando a tabela *lac* e os gráficos plotados, você ficou intrigado que o Chile (42º lugar) a Argentina (47º lugar), o Uruguai (58º Lugar) e o Peru (84º lugar) estavam acima do Brasil (87º lugar) na classificação do IDH. Para fazer outras análises, você deletou as colunas

```python

lac2 = lac.drop(columns=['ISO3', 'Human Development Groups', 'UNDP Developing Regions', 'HDI Rank (2021)'])
lac2
```

Para facilitar a proxima análise é interessante que a tabela seja transposta, ou seja, as linhas da tabela se tornem colunas e as colunas se tornem linhas.

Execute a célula abaixo para que esse procedimento seja executado. Não faça altearações nas declarações.

```python
# Execute esta célula sem alterar as declarações abaixo ########################
lac2.set_index('Country', inplace=True)
lac2 = lac2.transpose()
lac2.reset_index(inplace=True)
lac2.rename(columns={'index': 'Year'}, inplace=True)
lac2.columns.name = None
print('''
As 5 primeiras linhas da tabela lac2 são:
''')
lac2.head()
```


Agora você plotou o gráfico de linhas para analisar a evolução dos anos de estudos esperados no Chile, Argentina, Uruguai, Peru e Brasil de 1990 até 2021

```python
import matplotlib.pyplot as plt
paises_interesse = ['Chile', 'Argentina', 'Uruguay', 'Peru', 'Brazil']

df_interesse = lac2[['Year'] + paises_interesse]
plt.figure(figsize=(12, 8))

for pais in paises_interesse:
    plt.plot(df_interesse['Year'], df_interesse[pais], marker='o', label=pais)
plt.xlabel('Ano')
plt.ylabel('Expectativa em anos')
plt.title('Evolução da Expectativa de Anos de Estudo Formal (1990-2021)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()

```
